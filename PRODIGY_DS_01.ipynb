**1. Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/pop.csv")
df.head()
```

**2. Removing unwanted columns**

```python
new_df=df.drop(['Country Code','Indicator Name','Indicator Code'],axis=1)
```

```python
country='India'
country_info= new_df[new_df['Country Name']== country]
year=country_info.columns[2:].astype(int)
popul=country_info.iloc[:,2:].values.flatten()
```

**3. Plotting the Barchart**

```python
plt.figure(figsize=(5,3))
plt.bar(year, popul,color='red')
plt.xlabel("Year")
plt.ylabel('Population')
plt.title('Population of India over Years')
```

**4. Plotting the Histogram**

```python
plt.hist(popul,color='red',edgecolor='black')
plt.show()
```
